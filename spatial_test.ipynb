```python
import pytest
import ipytest
from spatial import *
```

```python
ipytest.autoconfig()
```

```python
%%ipytest

def test_to_vec_Nx3_broadcasting():
    # Test single vector to N rows
    vec = [0, 0, 1]
    result = to_vec_Nx3(3, vec)
    assert result.shape == (3, 3)
    assert np.all(result[2] == [0, 0, 1])

    # Test already correct shape
    existing = np.random.rand(5, 3)
    assert np.array_equal(to_vec_Nx3(5, existing), existing)

def test_to_scal_Nx1_logic():
    # Test scalar expansion
    assert to_scal_Nx1(4, 10.5).shape == (4, 1)
    assert np.all(to_scal_Nx1(4, 10.5) == 10.5)

    # Test None handling (default_val)
    res = to_scal_Nx1(2, None, default_val=1.0)
    assert np.all(res == 1.0)

def test_as_array_of_size():
    assert len(as_array_of_size(5, 10)) == 10
    with pytest.raises(ValueError):
        as_array_of_size([1, 2], 5)

def test_solar_panel_angular_efficiency():
    # Test normal incidence (cos(0) = 1) -> should be near 1.0
    assert 0.9 < solar_panel_angular_efficiency(np.array([1.0]))[0] <= 1.01
    # Test grazing/behind incidence (cos(90+) <= 0) -> should be 0
    assert solar_panel_angular_efficiency(np.array([0.0]))[0] == 0

def test_q_functions():
    # Q(0) should be 0.5
    assert Qfunction(0) == 0.5
    # Qinv(0.5) should be 0
    assert pytest.approx(Qinv(0.5)) == 0.0
```
